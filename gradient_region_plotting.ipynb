```python
import matplotlib.pyplot as plt
import numpy as np

def F1(x1, x2):
    if x1 + x2 > 2:
        return 1
    else:
        return 0

def F2(x1, x2):
    if x2 > 3:
        return 1
    else:
        return 0

def F3(x1, x2):
    if x1 > 3:
        return 1
    else:
        return 0

def F(x1, x2):
    if F1(x1, x2) == 1 and F2(x1, x2) == 1 and F3(x1, x2) == 1:
        return 1
    else:
        return 0

# Testing the functions with the given set of points
points = [(0, 0), (3.5, 3.5), (1.5, 1.5), (1.5, 3.5), (2.5, -0.25)]

# Plotting the decision lines and points
x = np.linspace(-1, 5, 100)
y1 = 2 - x
y2 = np.ones_like(x) * 3
y3 = np.linspace(-1, 5, 100)

fig, ax = plt.subplots()
ax.plot(x, y1, 'r', label='Line 1: x1 + x2 = 2')
ax.plot(x, y2, 'g', label='Line 2: x2 = 3')
ax.plot([3, 3], [-1, 5], 'b', label='Line 3: x1 = 3')
ax.fill_between(x, y1, where=(y1 > y2) & (y1 > y3), color='green', alpha=0.3)
ax.legend()

for point in points:
    x1, x2 = point
    label = f"({x1}, {x2})"
    ax.plot(x1, x2, 'ko', label=label)

plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Perceptron Decision Lines')
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def circle_boundary(x1, x2, radius, center):
    return np.where((x1 - center[0])**2 + (x2 - center[1])**2 <= radius**2, 1, 0)

def F(x1, x2):
    return circle_boundary(x1, x2, 2, (2, 2))

# Testing the function with the given set of points
points = [(0, 0), (3.5, 3.5), (1.5, 1.5), (1.5, 3.5), (2.5, -0.25)]

# Plotting the decision boundary and points
x = np.linspace(-3, 7, 200)
y = np.linspace(-3, 7, 200)
X, Y = np.meshgrid(x, y)
Z = F(X, Y)

fig, ax = plt.subplots()
ax.contourf(X, Y, Z, levels=[-0.5, 0.5, 1.5], colors=['white', 'green'], alpha=0.3)
ax.contour(X, Y, Z, levels=[-0.5, 0.5, 1.5], colors='blue')

for point in points:
    x1, x2 = point
    label = f"({x1}, {x2})"
    ax.plot(x1, x2, 'ko', label=label)

plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Perceptron Decision Boundary (Circle)')
plt.xlim(-3, 7)
plt.ylim(-3, 7)
plt.grid(True)
plt.legend()
plt.show()
```
